## Introduction

Die Untersuchung beschäftigt sich mit dem Thema Mobilität, insbesondere im Hinblick auf den öffentlichen Personenverkehr. Dieser umfasst die Grundversorgung von Mobilität auf Straße, Schiene, Wasser und mittels Luftseilbahn. Hierbei werden ausschließlich Daten von Fahrzeugen der Verkehrsbetriebe Zürich (VBZ) mit festen Soll-Abfahrtszeiten berücksichtigt. Dies beinhaltet sowohl Trams als auch Trolleybusse.
Das Thema Mobilität ist insbesondere im Hinblick auf aktuelle gesellschaftliche und politische Problemstellungen rund um die Erreichung von Klimazielen sehr relevant (Stückelberger, 2021).
Als “Welthauptstadt des ÖPNV” beschreibt Kühne (2018) die schweizer Stadt Zürich. Er verdeutlicht, dass 41% der Stadtbewohner:innen den öffentlichen Nahverkehr benutzen. Die Gründe dafür seien vielfältig: durch saubere Fahrzeuge, eine sehr gute Netzstruktur, Pünktlichkeit und einen abgestimmten Fahrplan werde eine zuverlässige Fortbewegung ermöglicht. 

Mit der Untersuchung setzt sich die Projektgruppe D das Ziel, dieses Angebot weiter zu verbessern. Mithilfe der Daten der Verkehrsbetriebe Zürich soll herausgefunden werden, ob es eine Korrelation zwischen der Auslastung im öffentlichen Personenverkehr und den Verspätungen gibt, um Handlungsempfehlungen zu bestimmten Auslastungszeiten abzuleiten.

Der Verkehrsverbund Nürnberg hat bereits Analysen zur Auslastung des ÖPNV unternommen und festgestellt, dass die Nutzung gegenüber der Vor-Corona Phasen zeitweise um 45% gesunken ist (Klodner, 2021). Bezogen auf Zürich soll dieser Vergleich ebenfalls analysiert werden um mit den Erkenntnissen folgende Hypothesen zu untersuchen:
1. Der Berufsverkehr wirkt sich negativ auf die Einhaltung der Soll-Fahrzeiten im öffentlichen Personenverkehr aus → Der öffentliche Verkehr ist zur Rush-hour überlastet
2. Während der Corona-Pandemie konnten die Soll-Zeiten besser eingehalten werden → Bei geringerer Auslastung ist die Pünktlichkeit sichergestellt
3. Linien die häufig in der Innenstadt verkehren haben ein höheres Risiko sich zu verspäten
4. Der Busverkehr ist unzuverlässiger als der Tramverkehr.
