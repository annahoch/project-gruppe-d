## Introduction

Die Untersuchung beschäftigt sich mit dem Thema Mobilität, insbesondere im Hinblick auf den öffentlichen Personenverkehr. Dieser umfasst die Grundversorgung von Mobilität auf Straße, Schiene, Wasser und mittels Luftseilbahn. Hierbei werden ausschließlich Daten von Fahrzeugen der Verkehrsbetriebe Zürich (VBZ) mit festen Soll-Abfahrtszeiten berücksichtigt. Dies beinhaltet sowohl Trams als auch Trolleybusse.
Das Thema Mobilität ist insbesondere im Hinblick auf aktuelle gesellschaftliche und politische Problemstellungen rund um die Erreichung von Klimazielen sehr relevant (Stückelberger, 2021).
Als “Welthauptstadt des ÖPNV” beschreibt Kühne (2018) die schweizer Stadt Zürich. Er verdeutlicht, dass 41% der Stadtbewohner:innen den öffentlichen Nahverkehr benutzen. Die Gründe dafür seien vielfältig: durch saubere Fahrzeuge, eine sehr gute Netzstruktur, Pünktlichkeit und einen abgestimmten Fahrplan werde eine zuverlässige Fortbewegung ermöglicht. 

Mit der Untersuchung setzt sich die Projektgruppe D das Ziel, dieses Angebot weiter zu verbessern. Mithilfe der Daten der Verkehrsbetriebe Zürich soll herausgefunden werden, ob es eine Korrelation zwischen der Auslastung im öffentlichen Personenverkehr und den Verspätungen gibt, um Handlungsempfehlungen zu bestimmten Auslastungszeiten abzuleiten.

Der Verkehrsverbund Nürnberg hat bereits Analysen zur Auslastung des ÖPNV unternommen und festgestellt, dass die Nutzung gegenüber der Vor-Corona Phasen zeitweise um 45% gesunken ist (Klodner, 2021). Bezogen auf Zürich soll dieser Vergleich ebenfalls analysiert werden um mit den Erkenntnissen folgende Hypothesen zu untersuchen:
1. Der Berufsverkehr wirkt sich negativ auf die Einhaltung der Soll-Fahrzeiten im öffentlichen Personenverkehr aus → Der öffentliche Verkehr ist zur Rush-hour überlastet
2. Während der Corona-Pandemie konnten die Soll-Zeiten besser eingehalten werden → Bei geringerer Auslastung ist die Pünktlichkeit sichergestellt
3. Linien die häufig in der Innenstadt verkehren haben ein höheres Risiko sich zu verspäten
4. Der Busverkehr ist unzuverlässiger als der Tramverkehr.



## Data description

Zur Analyse der Daten stehen wöchentliche csv-Dateien zur Verfügung, welche eine Vielzahl an Daten beinhalten. Jeder Eintrag stellt einen geplanten Halt eines Fahrzeugs der Verkehrsbetriebe Zürich dar und gibt eine präzise Auskunft über Datum, Linie, Fahrzeug, Haltestelle und vielem mehr. Somit ermöglicht der Datensatz einen Vergleich zwischen den geplanten und tatsächlich erfolgten Fahrzeiten der Fahrzeuge der Verkehrsbetriebe Zürich. Da vielen Daten aus der Tabelle eine interne ID und somit einen Fremdschlüssel zu anderen Tabellen darstellen, sind weitere Tabellen vorhanden, um diese Daten mit konkreten, verständlichen Bezeichnungen verknüpfen zu können.

Der Datenlieferant ist Verkehrsbetriebe Zürich, Departement der Industriellen Betriebe. Die Daten wurden erstmals 2015 erfasst und werden seither wöchentlich sonntags aktualisiert. Sämtliche Zeiten werden dabei auf die Sekunde gemessen. Über die genaue Bereitstellung der Daten gibt es keine Informationen. Da die Daten wöchentlich aktualisiert werden, ist davon auszugehen, dass die Daten durch einen automatischen Export aus einer oder mehreren weiteren Datenbanktabellen zusammengetragen und miteinander verknüpft werden.

Die zur Verfügung stehenden Datensätze sind sehr groß: Eine grobe Kalkulation ergab eine Anzahl von ca. 147 Mio. Datensätzen für einen Zeitraum von zwei Jahren. Aus diesem Grund erscheint es als sinnvoll, die Daten für die Auswertung herunterzubrechen. Dementsprechend werden jeweils die ersten Februarwochen der Jahre 2019 und 2021 miteinander verglichen. Eine weitere Konzentration auf ein oder zwei spezifische Linien (z.B. eine Bus- und eine Tramlinie) ist nicht ausgeschlossen.

## Analysis approach

Wie bereits in der Einführung erwähnt, setzt sich die Analyse mit der Frage auseinander, ob es eine Korrelation zwischen Auslastung und Verspätung gibt, um Handlungsempfehlungen abzuleiten. Auch wenn die Daten einen tiefen Einblick in jegliche Halte der Fahrzeuge der Verkehrsbetreibe Zürich ermöglichen, so fehlen dennoch Indikatoren für die tatsächliche Auslastung der Fahrzeuge, oder um diese ausrechnen zu können. Durch die Analyse des Verkehrsverbundes Nürnberg wurde allerdings festgestellt, dass die Nutzung der öffentlichen Verkehrsmittel während Corona im Vergleich zur Vor-Corona Zeit stark zurückgegangen ist. Diese Feststellung lässt sich als eine Hypothese auf die Stadt Zürich übertragen.

Geht man also davon aus, dass die Auslastung während Corona geringer war als vor Corona, so lassen sich mögliche Korrelationen zwischen Auslastung und Verspätung aufzeigen. Hierfür muss ein Vergleich der beiden Zeiträume miteinander vorgenommen werden. Im Handelsblatt verweist Blume darauf, dass einer der Lockdowns in der Schweiz von Mitte Januar bis Ende Februar 2021 verhängt wurde (2021), wodurch sich dieser Zeitraum gut dafür eignet, den Corona-Zeitraum zu repräsentieren. Auf dieser Grundlage wird die Analyse auf Februar 2019 und 2021 begrenzt.
Die Verspätung der einzelnen Fahrzeuge wird durch das Delta der jeweiligen Soll- und Ist-Wert errechnet.

Um eine gezielte Datenanalyse zu gewährleisten wird sich im Zuge der Auswertung für einen ersten Anlauf auf die folgende Variablen. Gegebenenfalls kann diese Liste erweitert oder reduziert werden:
- "linie": Interne ID der Linien; wird für eine Konzentration der Auswertung auf eine oder mehrere Linien benötigt
- "richtung": Gibt Auskunft über die Richtung, in welche sich das Fahrzeug bewegt; die Werte können dabei 1 oder 2 sein. Mit dieser Information können einzelne Fahrten rekonstruiert werden, z.B. um die Verspätungen an den hintereinander angefahrenen Haltestellen zu aggregieren.
- "betriebsdatum" und "datum_von": Tagesdatum des Haltestellenereignis des Betriebstags (05:00 bis 01.30 Uhr) & des tatsächlichen Tages.
- "fahrzeug": Interne Fahrzeug-ID
- "halt_diva_von" und "halt_punkt_diva_von": Interne IDs; geben Auskunft über die genaue Haltestelle und den genauen Haltepunkt eines Fahrzeugs
- "soll_an_von" und "ist_an_von": Soll- & Istankunft der Fahrzeuge in Sekunden nach Mitternacht
- "soll_ab_von" und "ist_ab_von": Soll- & Istabfahrt der Fahrzeuge in Sekunden nach Mitternacht

Die Werte einer groben deskriptiven Datenanalyse können in den nachfolgenden Codezeilen entnommen werden. Hierbei wir ein Überblick über die Anzahl der jeweiligen Datensätze, der minimal und maximal auftretenden Verspätungen bzw. Ankunftszeiten sowie die Summer dieser Zeiten. Ein genauerer Einblick inklusive Mittelwert (mean) und Median (50%) kann ebenfalls weiter unten vermittelt werden.

Um die Entscheidung für mögliche Visualisierungen der Daten zu treffen, wurden Aspekte herausgesucht, welche in einem späteren Dashboard dargestellt werden sollen. Im Folgenden werden diese Aspekte und deren mögliche Visualisierungen aufgezeigt:
- Gesamtbetrachtung des Verkehrsnetzes Zürich: Hier soll ein Bild Auskunft über die Linien in der Stadt geben und aufzeigen, auf welche Linien sich die Analyse konzentriert (z.B. Hauptverkehrslinien oder Linien nahe der Innenstadt)
- Verspätungen der Linien (aggregiert je Stunde oder Tag) im Berufsverkehr: Lollipop-Grafik
- Verspätungen der Linien (aggregiert je Stunde oder Tag) im Wochenendverkehr: Lollipop-Grafik
- Gesamtbetrachtung der Verspätung auf einer Linie: Connected Scatter (Y-Achse: Verspätung; X-Achse: Haltestellen nach Fahrplan)
- Verteilung der Verspätungen je Linie: Boxplot

### Code für die deskriptive Datenanalyse

```python
import pandas as pd

# reset display of max rows to enable displaying all columns of large dfs
pd.options.display.max_columns = None

# import csvs

# 1. Februarwoche 2019
df_19 = pd.read_csv("https://data.stadt-zuerich.ch/dataset/vbz_fahrzeiten_ogd_2019/download/Fahrzeiten_SOLL_IST_20190203_20190209.csv")

# 1. Februarwoche 2021
df_21 = pd.read_csv("https://data.stadt-zuerich.ch/dataset/vbz_fahrzeiten_ogd_2021/download/Fahrzeiten_SOLL_IST_20210131_20210206.csv")
```

```python
# Grober Überblick

# count rows
row_count_19 = len(df_19.index)
print("row count 19: ", row_count_19)
row_count_21 = len(df_21.index)
print("row count 21: ", row_count_21)
print("----------")

# add column verspaetung
df_19["verspaetung"] = df_19["ist_an_von"] - df_19["soll_an_von"]
df_21["verspaetung"] = df_21["ist_an_von"] - df_21["soll_an_von"]

# min verspaetung
min_19 = df_19["verspaetung"].min()
print("min verspaetung 19: ", min_19)
min_21 = df_21["verspaetung"].min()
print("min verspaetung 21: ", min_21)
print("----------")

# max verspaetung
max_19 = df_19["verspaetung"].max()
print("max verspaetung 19: ", max_19)
max_21 = df_21["verspaetung"].max()
print("max verspaetung 21: ", max_21)
print("----------")

# sum verspaetung
sum_19 = df_19["verspaetung"].sum()
print("sum verspaetung 19: ", sum_19)
sum_21 = df_21["verspaetung"].sum()
print("sum verspaetung 21: ", sum_21)
```

```python
# Genauerer Einblick (Mittelwert, Median, etc.)

# describe 19
print("describe 19:")
print(df_19["verspaetung"].describe())
# describe 21
print("describe 21:")
print(df_21["verspaetung"].describe())
```
